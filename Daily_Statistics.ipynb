```python
import pandas as pd
import numpy as np
df= pd.read_csv("Clean_DF.csv")
df
```

```python
df= df.rename(columns={'Date':'Day'})
dates = pd.to_datetime(df.iloc[:,0:5]).to_frame()
df = dates.join(df.iloc[:,5:])
df= df.rename(columns={0:'Date'})
```

```python
df=df.set_index('Date')
```

```python
df["Rainfall"].shift(-18)
left=df.iloc[:,:-1].resample('24H').mean() 
right=df.iloc[:,-1].resample('24H').max()
ndf=left.join(right)
ndf
```

```python
ndf.to_csv("df_daily_avg.csv", index=False)
```
